```python
# Imports
import RTModel
import os
import time
import RTModel.plotmodel as plm
```


# Running RTModel

```python
# Function to run RTModel
def run_rtmodel(event_path_, number_of_processors_=os.cpu_count()):
    start = time.time()
    # Call RTModel
    rtm = RTModel.RTModel()
    # Set the number of processors you'd like to use. If not specified, it will use all available processors.
    rtm.set_processors(number_of_processors_)
    rtm.set_event(event_path_)
    # Execute RTModel
    rtm.run(cleanup=True)

    print(f'Done!')
    finish = time.time()
    print('RTModel took', finish - start, f'seconds'
                                          f' using {number_of_processors_} processors.')
```

```python
run_rtmodel('mb20135_RTModel_run', number_of_processors_=14)
```
